## Research Question

What lifestyle factors (e.g., physical activity, stress, occupation, etc) are most strongly associated with sleep quality?

## Background and Prior Work

Sleep is an important factor in the human body’s overall health, and it is essential for all individuals to get sufficient sleep to support cognitive function, metabolism regulation, and productivity in day-to-day tasks. However, it is important to note that a lack of sleep in many adults can lead to severe health issues, such as a decreased immune response as well as disruptions in mood regulation and mental health. Because of the major role that sleep plays in overall well-being, researchers have studied how certain lifestyle and environmental factors influence sleep quality.


Researchers have found that sleep does not only pertain to how long one sleeps, but also the quality and duration of the sleep that occurs. For example, sleep must be uninterrupted and allow the body to repair and restore its functions overnight in order to prepare for the following day. With this in mind, prior research explains how good-quality sleep allows both the brain and body to recover, while poor sleep can carry similar risks to getting little to no sleep at all. According to the Sleep Foundation, insufficient or low-quality sleep is linked to weight gain, heart disease, stroke, and mental health problems such as anxiety and depression.<a name="cite_ref-SleepFoundation"></a><sup>1</sup>


In addition, it is important to note that researchers have studied various lifestyle factors that are linked to the quality and duration of sleep individuals are able to obtain. For instance, factors such as diet, including excessive junk food consumption, screen time, social media use, and physical activity have all been associated with sleep quality.<a name="cite_ref-Dzierzewski2021"></a><sup>2</sup> Researchers have also observed that regular physical activity throughout the day, such as a sufficient amount of cardiovascular exercise, can lead to better overall sleep patterns, especially as individuals age. These findings suggest that unhealthy habits can result in poorer sleep outcomes, while active lifestyles support better sleep quality.<a name="cite_ref-Dzierzewski2021"></a><sup>2</sup>


In a study conducted in China, it was found that a large population of undergraduate students experienced poor sleep quality due to low levels of physical exercise and high amounts of screen time before bedtime.<a name="cite_ref-Wu2022"></a><sup>3</sup> Although this project will examine several additional factors such as gender, age, occupation, BMI, and blood pressure in addition to lifestyle behaviors like physical exercise, it is important to recognize that each of these factors may have a significant impact on sleep quality. These relationships will be explored further throughout this project.


## Hypothesis


Hypothesis: Sleep quality will be worse among individuals with higher BMI or obesity, older age, and abnormal blood pressure. We also expect sleep quality to differ across occupations and gender, with people in more sedentary or high-stress jobs and men reporting poorer sleep overall.

Rationale: Sleep is influenced by both physical health and daily lifestyle. Higher BMI and obesity are linked to sleep problems like breathing disruptions, while aging is often associated with more fragmented sleep. Blood pressure can reflect stress and underlying health conditions that interfere with rest. Occupation and gender may also shape sleep through differences in stress levels, routines, and biological factors, making them important variables to examine when studying sleep quality.

## Data

The ideal dataset would contain clear, primarily numerical measurements of a person’s sleep quality and the lifestyle habits that are typically associated with sleep. These variables would include physical activity level, daily steps, stress level, heart rate, occupation, and more. This dataset should also take note of demographic characteristics such as age and gender, which can influence sleep quality as well. We would ideally want over 200 observations that have a balanced distribution of different ages, genders, and types of habits. A dataset without any biased representation would ensure that the results of our project are reliable and applicable to the real-world. 


Moreover, the dataset should include numerical data regarding the subjects’ personal and health indicators. Certain measurements such as stress level and sleep quality are highly subjective due to being self-reported, so these features would be difficult to analyze without a numerical representation. For example, one subject’s good quality of sleep could be another subject’s bad quality of sleep. Such subjectivity also amplifies the natural bias in the dataset. Therefore, the data should clearly define a numerical scale that can be more objectively interpreted and easily analyzed. 


The Kaggle Sleep, Health, and Lifestyle dataset is suitable for answering our research question because it contains several measurements of lifestyle habits, health, and demographics such as physical activity levels, stress levels, BMI categories, heart rate, blood pressure, age, and gender. This dataset also contains useful sleep metrics such as sleep quality, sleep disorders, and sleep duration. This dataset also contains 400 observations that are widely spread across ages 27-59 and have an even distribution of male and female subjects. All features are numerically represented, making them easy to analyze and draw conclusions from. In addition, the Kaggle Sleep Efficiency dataset can be used as a supporting dataset to analyze sleep quality at a deeper level using efficiency and sleep stage measurements, while the primary dataset focuses on broader lifestyle habits. This secondary dataset also provides bedtime and wakeup time habits, which can also be studied to support our analysis. Collectively, both of these datasets can help us answer our research question.
